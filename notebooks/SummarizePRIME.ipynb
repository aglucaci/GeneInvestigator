```python
import pandas as pd
import numpy as np
import os
import json
import altair as alt
import sys
```

```python

JSON_FILE = "../results/BDNF/Recombinants/BDNF_codons_RDP_recombinationFree.fas.PRIME.json"
pvalueThreshold = 0.1
```

```python
# "headers"

def getPRIME_Headers(json_file):
    with open(json_file, "r") as in_d:
        json_data = json.load(in_d)
    return json_data["MLE"]["headers"]
#end method

def getPRIME_Data(json_file):
    with open(json_file, "r") as in_d:
        json_data = json.load(in_d)
    return json_data["MLE"]["content"]["0"]
#end method
```

```python
columns = getPRIME_Headers(JSON_FILE)
headers = [x[0] for x in columns]
headers
```

```python
atchley_symbols=['L1: Polarity index', 'L2: Secondary structure factor', 'L3: Volume', 'L4: Refractivity/Heat Capacity', 'L5: Charge/Iso-electric point']
```

```python
df = pd.DataFrame(getPRIME_Data(JSON_FILE), columns=headers, dtype = float)
df.index += 1
df["Site"] = df.index
df
```

```python
overall_pvalue_df = df[df["p-value"] <= 0.1]
overall_pvalue_df
```

```python
#overall_pvalue_df["Site"].plot()
# f.plot.bar(x='lab', y='val', rot=0)
overall_pvalue_df["Site"].plot.bar(x="Site", y="p-value")
```

```python
p1_pvalue_df = df[df["p1"] <= pvalueThreshold]
p1_pvalue_df
```

```python
p2_pvalue_df = df[df["p2"] <= pvalueThreshold]
p2_pvalue_df
```

```python
p3_pvalue_df = df[df["p3"] <= pvalueThreshold]
p3_pvalue_df
```

```python
p4_pvalue_df = df[df["p4"] <= pvalueThreshold]
p4_pvalue_df
```

```python
p5_pvalue_df = df[df["p5"] <= pvalueThreshold]
p5_pvalue_df
```
