```python
import pandas as pd
import numpy as np
import os
import json
import altair as alt
```

```python
#JSON_FILE = "../results/TP53/TP53_codons.fasta.FMM.json"
JSON_FILE = "../results/BDNF/Recombinants/BDNF_codons_RDP_recombinationFree.fas.FMM.json"
pvalueThreshold = 0.1
```

```python
def getData(json_file):
    with open(json_file, "r") as in_d:
        json_data = json.load(in_d)
    return json_data["Evidence Ratios"]
#end method

def getFits(json_file):
    with open(json_file, "r") as in_d:
        json_data = json.load(in_d)
    return json_data["fits"]
#end method
```

```python
json_file = JSON_FILE
with open(json_file, "r") as in_d:
    json_data = json.load(in_d)
    
print(json_data.keys())
print(json_data["Evidence Ratios"].keys())
```

```python
df = ""
for key in json_data["Evidence Ratios"].keys():
    print(key)
    if type(df) != str:
        df2 = pd.DataFrame.from_dict(getData(JSON_FILE)[key]).T
        df2 = df2.rename(columns={0:key})
        df2[key] = 2 * np.log(df2[key])
        df[key] = df2[key]
    else:
        df = pd.DataFrame.from_dict(getData(JSON_FILE)[key]).T
        df = df.rename(columns={0:key})
        df[key] = 2 * np.log(df[key])
    #end if
    #print(df)
#end for

df.index += 1
df["Site"] = df.index
df
```

```python
import altair as alt

source = df

line = alt.Chart(source).mark_area().encode(
    x='Site',
    y='Triple-hit vs Double-hit'
)

line 
```
